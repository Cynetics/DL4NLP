```python
from collections import defaultdict
from pprint import pprint
def get_breakdown_first100(lijstje):
    breakdown = defaultdict(int)

    for z in lijstje[:100]:
        if z in anna_labels:
            for char in anna_labels[z]:
                breakdown[char] += 1
        else:
            breakdown['unk'] += 1
    return sorted(breakdown.items(), key = lambda x:x[1] ,reverse=True)

print("Socrates datapoints")
print('baseline', get_breakdown_first100(socrates_baseline))
print('chars', get_breakdown_first100(socrates_chars))
print('bytes', get_breakdown_first100(socrates_bytes))


print("")
print("Uncertain datapoints")
print('baseline', get_breakdown_first100(uncertain_base))
print('chars', get_breakdown_first100(uncertain_chars))
print('bytes', get_breakdown_first100(uncertain_bytes))
```

```python
for a in [bay_baseline, bay_chars, bay_bytes]:
    total = 0 
    lang_socrates = dict()
    lang_uncertainties = dict()
    for lan in all_lans:
        sds, amt_socr = a.language_stats(lan, plot=False)
        lang_uncertainties[lang_dict[lan]] = sds
        lang_socrates[lang_dict[lan]] = amt_socr
        
        total += sds
    print(total, total/len(all_lans))
    print(list(sorted(lang_uncertainties.items(), key=lambda x:x[1], reverse=True))[:15])
    print(list(sorted(lang_socrates.items(), key=lambda x:x[1], reverse=True))[:15])
```

```python
def get_sds(a, threshold=0.002):
    sds = [np.mean(a.results_dict[z]['sds']) for z in a.results_dict]
    return [z for z in sds if z > threshold]
sds_data = [get_sds(bay_baseline), get_sds(bay_chars), get_sds(bay_bytes)]
```

```python
softmaxes_data = [bay_baseline.top_softmaxes(), bay_chars.top_softmaxes(), bay_bytes.top_softmaxes()]
```
